```python
rank = pd.read_csv('./scrapy_csv_files/glassdoor_ranks/rank.csv')
```

```python
rank
```

Manually fix incorrectly labeled cells

```python
rank.loc[10,'year'] = float(2019)
rank.loc[0,'year'] = float(2018)
rank.loc[1,'year'] = float(2017)
rank = rank.sort_values('year',ascending = False)
```

```python
rank = rank.drop(['end_date', 'start_date'], axis = 1)
```

```python
rank = rank.reset_index().drop(['index'], axis =1)
```

```python
co = rank.company_list.str.split(',',expand= True)
co.iloc[0,20] = 'Ryan LLC'
```

```python
co
```

Transpose

```python
co = rank.company_list.str.split(',',expand= True)
co.iloc[0]=co.iloc[0].drop([21], axis =0)
co.iloc[0,20] = 'Ryan LLC'
co.iloc[0] = co.iloc[0].dropna().reset_index().drop('index', axis = 1)[0]
co = co.T
co.columns = rank.year
co
```

Unpivot table

```python
com = pd.melt(co, var_name = 'year', value_name = 'company')
com = com.dropna().reset_index().drop('index', axis = 1)
com
```

```python
cr = rank.company_rank.str.split(',',expand= True)
cr = cr.T
cr.columns = rank.year
cr
```

```python
crm = pd.melt(cr, var_name = 'year', value_name = 'rank')
crm = crm.dropna().reset_index().drop('index', axis = 1)
crm
```

```python
cs = rank.company_score.str.split(',',expand= True)
cs = cs.T
cs.columns = rank.year
cs
```

```python
csm = pd.melt(cs, var_name = 'year', value_name = 'score')
csm = csm.dropna().reset_index().drop('index', axis = 1)
csm
```

```python
rank_final = pd.concat([crm,com,csm],axis =1)
```

```python
rank_final.columns = ['year','rank','year1','company','year2','score']
```

```python
rank_final = rank_final.drop(['year1','year2'],axis = 1)
```

```python
rank_final
```

```python
rank_final.to_csv('./scrapy_csv_files/glassdoor_ranks/rank_final.csv')
```
